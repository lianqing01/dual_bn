```python
import torch
from models.constraint_bn_v2 import *
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
net = torch.load("epoch_99.pth")
net_119 = torch.load("epoch_119.pth")
```

```python
layer = 0
for m,n in zip(net.modules(), net_119.modules()):
    if isinstance(m, Constraint_Norm):
        layer+=1
        mu = m.noise_mu_sum
        gamma_noise = m.nosie_gamma_sum
        mu_119 = n.noise_mu_sum
        gamma_119_noise = n.nosie_gamma_sum
        gamma = m.gamma_
        gamma_119 = n.gamma_
        for i in tqdm(range(m.num_features)):
            plt.figure(figsize=(10,10))
            plt.subplot(221)
            plt.hist(mu[:,:,i].cpu().numpy().reshape(391,-1), bins=25)
            plt.title("mu_epoch_{}_layer{}_index{}".format(99, layer, i))
            plt.subplot(222)

            plt.hist(mu_119[:,:,i].cpu().numpy().reshape(391,-1), bins=25)
            plt.title("mu_epoch_{}_layer{}_index{}".format(119, layer, i))
            plt.subplot(223)

            plt.hist(((gamma[:,:,i] - gamma_noise[:,:,i])**2).detach().cpu().numpy().reshape(391,-1), bins=25)
            plt.title("gamma_epoch_{}_layer{}_index{}".format(99, layer, i))
            plt.subplot(224)

            plt.hist(((gamma_119[:,:,i] - gamma_119_noise[:,:,i])**2).detach().cpu().numpy().reshape(391,-1), bins=25)
            plt.title("gamma_epoch_{}_layer{}_index{}".format(119, layer, i))
            plt.show()
            print(111)
            import pdb
            pdb.set_trace()
```
