```python
net = torch.load("epoch_99_noise_after_x.pth", map_location="cpu")
net_119 = torch.load("epoch_109_noise_after_x.pth", map_location="cpu")
```

```python
layer = 0
for m,n in zip(net.modules(), net_119.modules()):
    if isinstance(m, Constraint_Norm):
        layer+=1
        if layer < 10:
            continue
        x = torch.cat(m.summarize_x_hat, dim=0).cpu()[:128]
        x_noise = torch.cat(m.summarize_x_hat_noise, dim=0).cpu()[:128]

        x = x.mean(2).mean(2)
        x_noise = x_noise.mean(2).mean(2)

        x_109 = torch.cat(n.summarize_x_hat, dim=0).cpu()[:128]
        x_noise_109 = torch.cat(n.summarize_x_hat_noise, dim=0).cpu()[:128]

        x_109 = x_109.mean(2).mean(2)
        x_noise_109 = x_noise_109.mean(2).mean(2)
  
        for i in tqdm(range(m.num_features)):
            plt.figure(figsize=(10,10))
            plt.subplot(221)
            plt.hist(x[:,i], bins=25)
            plt.title("x_layer_{}".format(99, layer))
            plt.subplot(222)

            plt.hist(x_noise[:,i], bins=25)
            plt.title("x_layer_noise_{}".format(99, layer))
            plt.subplot(223)
            plt.hist(x_109[:,i], bins=25)
            plt.title("x_layer_{}".format(119, layer))
            plt.subplot(224)

            plt.hist(x_noise_109[:,i], bins=25)
            plt.title("x_layer_noise_{}".format(119, layer))
            plt.show()
            import pdb
            pdb.set_trace()
            
```
